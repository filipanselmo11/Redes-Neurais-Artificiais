### Função de ativação sigmoide

$$f(x) = \frac{1}{1 + e^{-\beta \cdot x}}$$

- Implemente uma função em Python que retorne o valor da função sigmóide para uma dada entrada
- Nesta função, considere beta como parâmetro nomeado com valor igual a 1.0
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função sigmoide
- Em um mesmo gráfico, ilustre 3 valores distintos para $\beta$:
  - $\beta = 0.7$ na cor laranja
  - $\beta = 0.5$ na cor verde
  - $\beta = 0.3$ na cor vermelha
  - Use legendas no gráfico

```python
def sigmoid(x, beta=1.0):
    return (1/(1+np.exp(-beta*x)))
x = entrada
z = sigmoid(x)
#z = 1/(1 + np.exp(-x)) 
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)")

plt.title("Função de ativação sigmoide")
plt.show() 
```

```python
red = sigmoid(x, 0.3)
green = sigmoid(x, 0.5)
orange = sigmoid(x, 0.7)
#z = 1/(1 + np.exp(-x)) 

plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 

plt.plot(x, orange, color='orange', label="b = 0.7")
plt.plot(x, green, color='green', label="b = 0.5")
plt.plot(x, red, color='red', label="b = 0.3")
plt.legend(loc="upper left")
plt.show() 
```

```python

#Derivada sigmoide
def dersigmoid(x,beta=1.0):
    return (beta*np.exp(-beta*x)/(1 + np.exp(-beta*x))**2)

plt.plot(x, dersigmoid(x))
plt.title("Derivada da sigmoide")
plt.show() 
```

### Função de Ativação ReLU (Unidade Linear Retificada)

$$f(x) = max(0,x)$$

- Implemente uma função em Python que retorne o valor da função ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função ReLU

```python
def relu(x):
    try: # A entrada é um array
        arr = np.array([])
        for element in x:
            if(element > 0):
                arr = np.append(arr,element,axis=None)
            else:
                arr = np.append(arr,0,axis=None)
        return arr
    except: # A entrada não é um array
        if (x > 0):
            return x
        else:
            return 0
y = relu(x)
plt.plot(x, y)
plt.title("Função de ativação ReLU")
plt.show() 
```

```python
def derrelu(x):
    try: # A entrada é um array
        arr = np.array([])
        for element in x:
            if(element > 0):
                arr = np.append(arr,1,axis=None)
            else:
                arr = np.append(arr,0,axis=None)
        return arr
    except: # A entrada não é um array
        if (x > 0):
            return 1
        else:
            return 0

y = derrelu(x)
plt.plot(x, y)
plt.title("Derivada da ReLU")
plt.show() 
```

### Função de ativação linear (purelin)

$$f(x) = x$$

- Implemente uma função em Python que retorne o valor da função linear para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão

```python
def purelin(x):
    return x
    
plt.plot(x, purelin(x))
plt.title("Purelin")
plt.show() 
```

```python
def derpurelin(x):
    # Derivada de x é sempre 1!
    try: # A entrada é um array
        arr = np.array([])
        for element in x:
            arr = np.append(arr,1,axis=None) 
        return arr # Retorna um array só de 1
    except: # A entrada não é um array
        return 1 # Retorna 1!

plt.plot(x, derpurelin(x))
plt.title("Derivada da Purelin")
plt.show() 
```

### Função de ativação Tangente Hiperbólica

- Implemente uma função que retorne o valor da derivada da função em questão

```python
def tanh(x):
    return np.tanh(x)

plt.plot(x, tanh(x))
plt.title("Tangente hiperbólica")
plt.show() 
```

```python
def dertanh(x):
    return(1/(np.cosh(x)**2))

x2 = np.arange(-1,1,0.1)
plt.title("Derivada da Tangente hiperbólica")
plt.plot(x2, dertanh(x2))
plt.show()
```

### Função de Ativação Gaussiana

$$f(x) = \exp \left( \frac{-(x - \mu)^2}{2\sigma^2} \right)$$

- $\mu$ é o ponto médio (centro) da curva
- $\sigma$ denota o raio de abertura
- Implemente uma função em Python que retorne o valor da função gaussiana para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão


Considere também os seguintes valores ao plotar a função

- Parâmetros nomeados $\mu = 0$ e $\sigma = 1$
- $\mu = 0, \sigma = 0.3$
- $\mu = 0, \sigma = 0.5$
- $\mu = 0, \sigma = 0.7$
- $\mu = 0, \sigma = 0.9$
- $\mu = 0, \sigma = 1.1$

```python
def gaussian(x, mi, sigma):
    return (np.exp((-(x-mi)**2)/(2*(sigma**2))))

def dergaussian(x, mi, sigma):
    return (((-x-mi)/(sigma**2))*gaussian(x,mi,sigma))
```

```python
plt.plot(x, gaussian(x, 0,1),color='blue',label="sigma=1")
plt.plot(x, gaussian(x, 0,0.3),color='red',label="sigma=0.3")
plt.plot(x, gaussian(x, 0,0.5),color='black',label="sigma=0.5")
plt.plot(x, gaussian(x, 0,0.7),color='green',label="sigma=0.7")
plt.plot(x, gaussian(x, 0,0.9),color='orange',label="sigma=0.9")
plt.plot(x, gaussian(x, 0,1.1),color='pink',label="sigma=1.1")
plt.legend(loc="upper left")
plt.show() 
```

### Função de Ativação Leaky ReLU

$$f(x) = \left\{\begin{array}{cc} 0.01\cdot x & x < 0\\
x & x \geq 0
\end{array}
\right.$$

- Implemente uma função em Python que retorne o valor da função Leaky ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função Leaky ReLU

```python
def leakyrelu(x):
    try:
        # Se a entrada for um array
        arr = np.array([])
        for element in x:
            if(element > 0):
                arr = np.append(arr,element,axis=None)
            else:
                arr = np.append(arr,0.01*element,axis=None)
        return arr
    except:
        # Se a entrada não for um array
        if (x > 0):
            return x
        else:
            return 0.01*x
        
    
plt.plot(x, leakyrelu(x))
plt.title("Leaky ReLU")
plt.show() 
```

```python
def derleakyrelu(x):
    arr = np.array([])
    for element in x:
        if(element > 0):
            arr = np.append(arr,1,axis=None)
        else:
            arr = np.append(arr,0.01,axis=None)
    return arr


plt.plot(x, derleakyrelu(x))
plt.title("Derivada da Leaky ReLU")
plt.show()
```
