```python
#Célula reservada para importar as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-4, 4, 0.1)
print(x)
```

```python
print("Tipo de dado: ",(x.dtype))
```

```python
#f(x) = 1/(1 + e^-x)
def sigmoide(x,beta = 1.0):
    resultado = 1/(1 + np.exp(-beta*x))
    return resultado

def dsigmoide(x,beta=1.0):
    resultado = beta * np.exp(-beta * x) / (1 + np.exp(-beta * x)) ** 2
    return resultado
```

```python
plt.plot(x, sigmoide(x))
plt.plot(x, sigmoide(x,0.3), color='orange', label="beta = 0.7")
plt.plot(x, sigmoide(x,0.5), color='green', label="beta = 0.5")
plt.plot(x, sigmoide(x,0.7), color='red', label="beta = 0.3")

plt.legend(loc="upper left")
plt.xlabel("x")
plt.ylabel("Sigmoide(x)")
plt.title("Função de Ativação sigmoide")
plt.show()
```

```python
plt.plot(x, dsigmoide(x))
plt.plot(x, dsigmoide(x, 0.7), color='orange', label='beta=0.7')
plt.plot(x, dsigmoide(x, 0.5), color='green', label='beta=0.5')
plt.plot(x, dsigmoide(x, 0.3), color='red', label='beta=0.3')

plt.title("Derivada da função sigmoide")
plt.show()
```

```python
#Função de ativação RELU e sua derivada
def relu(x):
    try:
        arr = np.array([])
        for i in x:
            if(i > 0):
                arr = np.append(arr, i, axis=None)
            else:
                arr = np.append(arr, 0, axis=None)
        return arr
    except:
        if(x > 0):
            return x
        else:
            return 0

def d_relu(x):
    try:
        arr = np.array([])
        for i in x:
            if(i > 0):
                arr = np.append(arr, 1, axis=None)
            else:
                arr = np.append(arr, 0, axis=None)
        return arr
    except:
        if(x > 0):
            return 1
        else:
            return 0
```

```python
plt.plot(x, relu(x))
plt.title("Função de Ativação RELU")
plt.show()
```

```python
plt.plot(x, d_relu(x))
plt.title("Derivada da função RELU")
plt.show()
```

```python
def linear(x):
    return x

def d_linear(x):
    try: 
        arr = np.array([])
        for element in x:
            arr = np.append(arr,1,axis=None) 
        return arr
    except: 
        return 1 
```

```python
plt.plot(x, linear(x))
plt.title("Função de Ativação Linear")
plt.show()
```

```python
plt.plot(x, d_linear(x))
plt.title("Derivada da Função Linear")
plt.show()
```

```python
def tangente_hiperbolica(x):
    return np.tanh(x), x

def d_tangente_hiperbolica(x):
    return 1 - np.square(x)
```

```python
x_th, x = tangente_hiperbolica(x)
x_intervalo = x[(x >= -1.0) & (x < 1.1)]

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, x_th, 'r', label="Tangente Hip")
plt.plot(x, d_tangente_hiperbolica(x_th), 'y', label="Derivada")
plt.legend(loc="upper left")
plt.xlabel("x")
plt.ylabel(r"$\frac{e^x - e^{-x}}{e^x + e^{-x}}$")
plt.title("Função de Ativação Tangente Hiperbólica", fontsize=15)

plt.subplot(2, 2, 2)
x2_th, x2 = tangente_hiperbolica(x_intervalo)
plt.plot(x2, x2_th, 'r', label="Tangente Hiperbólica")
plt.plot(x2, d_tangente_hiperbolica(x2_th), 'y', label="Derivada")
plt.legend(loc="lower right")
plt.xlabel("x")
plt.title("Função de Ativação Tangente Hiperbólica", fontsize=15)

plt.show()
```

```python
def gaussiana(x, mi, sig):
    return np.exp(-np.power(x - mi, 2.) / (2 * np.power(sig, 2.)))

def d_gaussiana(x, mi, sig):
    return (((-x - mi) / (sig**2)) * gaussiana(x, mi, sig))
```

```python
plt.plot(x, gaussiana(x, 0,1), color='blue', label="sigma=1")
plt.plot(x, gaussiana(x, 0,0.3), color='red', label="sigma=0.3")
plt.plot(x, gaussiana(x, 0,0.5), color='black', label="sigma=0.5")
plt.plot(x, gaussiana(x, 0,0.7), color='green', label="sigma=0.7")
plt.plot(x, gaussiana(x, 0,0.9), color='orange', label="sigma=0.9")
plt.plot(x, gaussiana(x, 0,1.1), color='pink', label="sigma=1.1")
plt.legend(loc="upper left")
plt.title("Função de Ativação Gaussiana", fontsize=15)
plt.show()
```

```python
plt.plot(x, d_gaussiana(x, 0,1), color='blue', label="sigma=1")
plt.plot(x, d_gaussiana(x, 0,0.3), color='red', label="sigma=0.3")
plt.plot(x, d_gaussiana(x, 0,0.5), color='black', label="sigma=0.5")
plt.plot(x, d_gaussiana(x, 0,0.7), color='green', label="sigma=0.7")
plt.plot(x, d_gaussiana(x, 0,0.9), color='orange', label="sigma=0.9")
plt.plot(x, d_gaussiana(x, 0,1.1), color='pink', label="sigma=1.1")
plt.legend(loc="upper right")
plt.title("Função de Ativação Gaussiana Derivada", fontsize=15)
plt.show()
```

```python
def leaky_relu(x):
    try:
        arr = np.array([])
        for element in x:
            if(element > 0):
                arr = np.append(arr,element,axis=None)
            else:
                arr = np.append(arr,0.01*element,axis=None)
        return arr
    except:
        if (x > 0):
            return x
        else:
            return 0.01*x

def d_leaky_relu(x):
    arr = np.array([])
    for element in x:
        if(element > 0):
            arr = np.append(arr,1,axis=None)
        else:
            arr = np.append(arr,0.01,axis=None)
    return arr
```

```python
plt.plot(x, leaky_relu(x))
plt.title("Função de Ativação Leaky RELU")
plt.show()
```

```python
plt.plot(x, d_leaky_relu(x))
plt.title("Derivada da Função Leaky ReLU")
plt.show()
```
