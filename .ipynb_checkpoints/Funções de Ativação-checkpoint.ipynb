```python
#Célula reservada para importar as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-4, 4, 0.1)
print(x)
```

```python
print("Tipo de dado: ",(x.dtype))
```

```python
print("Quantidade de amostras: ", x.shape[0])
```

```python
#f(x) = 1/(1 + e^-x)
def sigmoide(x,beta):
    resultado = 1/(1 + np.exp(-beta*x)), x
    return resultado

def dsigmoide(x):
    resultado = sigmoide(x,beta) * (1 - sigmoide(x,beta)), x
    return resultado
```

```python
xSigmoide, x = sigmoide(x, 1.0)
aSigmoide, a = sigmoide(x, 0.7)
bSigmoide, b = sigmoide(x, 0.5)
cSigmoide, c = sigmoide(x, 0.3)

plt.plot(x, xSigmoide, label="Sigmoide")
plt.plot(a,dsigmoide(aSigmoide), label="Derivada \u03B2: 0.7")
plt.plot(b,dsigmoide(bSigmoide), label="Derivada \u03B2: 0.5")
plt.plot(c,dsigmoide(cSigmoide), label="Derivada \u03B2: 0.3")
plt.legend(loc="upper left")
plt.xlabel("x")
plt.ylabel(r"$\frac{1}{1 + e^{-beta.x}}$")
plt.title("Função de ativação sigmoide", fontsize=15)
plt.show()
```

```python
#Função de ativação RELU e sua derivada
def relu(x):
    resultado = np.maximum(0,x), x
    return resultado

def d_relu(x):
    if(x.any() <= 0):
        resultado = 0
    elif(x.any() > 0):
        resultado = 1
    return resultado, x
```

```python
retaRelu, x = relu(x)

plt.plot(x, retaRelu, 'b', label="RELU")
plt.plot(a, d_relu(retaRelu), 'r', label="Derivada")
plt.xlabel("x")
plt.ylabel(r"$max\{0, x\}$")
plt.legend(loc="upper left")
plt.title("Função de Ativação RELU", fontsize=15)
plt.show()
```

```python
def linear(x):
    return x, x

def d_linear(x):
    return 1,x
```

```python
retaL, x = linear(x)
retaDerivadaL = d_linear(x)

plt.plot(x, retaL, 'r', label="Linear")
plt.axhline(y=retaDerivadaL, xmin=x[0], xmax=x[-1], color = 'y', label="Derivada")
plt.legend(loc="upper left")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Função de Ativação Linear", fontsize=15)
plt.show()
```

```python
def tangente_hiperbolica(x):
    return np.tanh(x), x

def d_tangente_hiperbolica(x):
    return 1 - np.square(x)
```

```python
x_th, x = tangente_hiperbolica(x)
x_intervalo = x[(x >= -1.0) & (x < 1.1)]

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, x_th, 'r', label="Tangente Hip")
plt.plot(x, d_tangente_hiperbolica(x_th), 'y', label="Derivada")
plt.legend(loc="upper left")
plt.xlabel("x")
plt.ylabel(r"$\frac{e^x - e^{-x}}{e^x + e^{-x}}$")
plt.title("Função de Ativação Tangente Hiperbólica", fontsize=15)

plt.subplot(2, 2, 2)
x2_th, x2 = tangente_hiperbolica(x_intervalo)
plt.plot(x2, x2_th, 'r', label="Tangente Hiperbólica")
plt.plot(x2, d_tangente_hiperbolica(x2_th), 'y', label="Derivada")
plt.legend(loc="lower right")
plt.xlabel("x")
plt.title("Função de Ativação Tangente Hiperbólica", fontsize=15)

plt.show()
```

```python
def gaussiana(x, mi, sig):
    return np.exp(-np.power(x - mi, 2.) / (2 * np.power(sig, 2.)))

def d_gaussiana(x, mi, sig):
    return (((-x - mi) / (sig**2)) * gaussiana(x, mi, sig))
```

```python
plt.plot(x, gaussiana(x, 0,1), color='blue', label="sigma=1")
plt.plot(x, gaussiana(x, 0,0.3), color='red', label="sigma=0.3")
plt.plot(x, gaussiana(x, 0,0.5), color='black', label="sigma=0.5")
plt.plot(x, gaussiana(x, 0,0.7), color='green', label="sigma=0.7")
plt.plot(x, gaussiana(x, 0,0.9), color='orange', label="sigma=0.9")
plt.plot(x, gaussiana(x, 0,1.1), color='pink', label="sigma=1.1")
plt.legend(loc="upper left")
plt.title("Função de Ativação Gaussiana", fontsize=15)
plt.show()
```

```python
plt.plot(x, d_gaussiana(x, 0,1), color='blue', label="sigma=1")
plt.plot(x, d_gaussiana(x, 0,0.3), color='red', label="sigma=0.3")
plt.plot(x, d_gaussiana(x, 0,0.5), color='black', label="sigma=0.5")
plt.plot(x, d_gaussiana(x, 0,0.7), color='green', label="sigma=0.7")
plt.plot(x, d_gaussiana(x, 0,0.9), color='orange', label="sigma=0.9")
plt.plot(x, d_gaussiana(x, 0,1.1), color='pink', label="sigma=1.1")
plt.legend(loc="upper right")
plt.title("Função de Ativação Gaussiana Derivada", fontsize=15)
plt.show()
```

```python
def leaky_relu(x):
    return np.maximum(0.1 * x, x), x

def d_leaky_relu(x):
    z = np.ones_like(x)
    z[x < 0] = 0.01
    return z, x
```

```python
retaLR, x = leaky_relu(x)
retaDerivadaLR, zx = d_leaky_relu(x)

plt.plot(x, retaLR, 'r', label="Leaky RELU")
plt.plot(zx, retaDerivadaLR, 'y', label="Derivada")
plt.xlabel("x")
plt.legend(loc="upper left")
plt.ylabel(r"$max\{0.1x, x\}$")
plt.title("Função de Ativação Leaky RELU", fontsize=15)
plt.show()
```
