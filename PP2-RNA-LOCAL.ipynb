## Algoritmo de Treinamento do Perceptron

### A célula abaixo contém, a codificação da função degrau do treinamento do neurônio perceptron, onde u é o paramêtro

```python
def step_function(u):
    if(u >= 0):
        return 1
    return 0
```

### A célula abaixo contém a codificação do treinamento do neurônio perceptron, onde N é a taxa de aprendizado, e é a taxa de erro, W são os pesos de entrada e yd é o Y desejado

```python
def perceptron(N, e, W):
    epoch = 1
    iscomplet = 0
    update = 0
    
    while(iscomplet != 1000):
        
        print("Época =========> ", epoch)
        print("Vetor de peos: ", W)
        
        islearn = 0
        wrong = 0
        
        for i in range(1000):
            
            xi = X[i]
            
            yd = Y[i]
            
            u = np.dot(xi, W)
            
            y = step_function(u)
            
            e = yd - y
            
            W = W + N*e*xi
            
            if(not e):
                islearn+= 1
            else:
                wrong += 1
                update += 1
                
        print("Exemplos corretos: ", islearn)
        print("Exemplos incorretos: ", wrong)
        
        iscomplet = islearn
        epoch += 1
        
        print("\n\n")
        
    return (W, epoch, update)
            
```
